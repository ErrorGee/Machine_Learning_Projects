```python
import ast
```

```python
def clean_genre(field):
    field=ast.literal_eval(field)
    for i in range(len(field)):
        #print(field[i])
        for j in field[i].values():
            field[i]=j
    return field

genre=clean_genre(field)
print(genre)
```

```python
movies["genres"]=movies["genres"].apply(clean_genre)
```

```python
cast=movies.iloc[0]["cast"]
```

```python
def clean_castcrew(cast):
    cast=ast.literal_eval(cast)
#     print(cast)
    counter=0
    for i in range(len(cast)):
        if counter!=3:
            cast[i]=cast[i]["name"]
            counter+=1
        else:
            break
    return cast[0:3]

topthree=clean_castcrew(cast)
print(topthree)
```

```python
movies["cast"]=movies["cast"].apply(clean_castcrew)
```

```python
movies.head(2)
```

```python
crew=movies.iloc[0]["crew"]
```

```python
movies.iloc[1]["crew"]
```

```python
def getDirector(crew):
#     print(crew)
    director_name=[]
    crew=ast.literal_eval(crew)
    for i in range(len(crew)):
        if crew[i]["job"]=="Director":
            director_name.append(crew[i]["name"])
            break
    return director_name
            
            
Director=getDirector(crew)
print(Director)
```

```python
movies["crew"]=movies["crew"].apply(getDirector)
```

```python
#lets do stemming first
!pip install nltk
```

```python
from  nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemit(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movie_df.iloc[0]["tags"]
```

```python
txt=movie_df.iloc[0]["tags"]
print(stemit(txt))
```

```python
movie_df["tags"]=movie_df["tags"].apply(stemit)
```

```python
movie_df.head(3)
```

Now Vectorization


```python
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors=CV.fit_transform(movie_df["tags"]).toarray()
```

```python
vectors.shape
```

```python
# CV.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

#the distance of every movie with every other movie
#we have 4806 movies, so the final shape should be 4806X4806
similarity=cosine_similarity(vectors)
```

```python
similarity.shape 
```

```python
movie_df[movie_df["title"]=="Batman"].index[0]
```

```python
# sorted(enumerate(similarity[0]), reverse=True, key=lambda x:x[1])
```

```python
def recommendation_system(movie):
    movie_index=movie_df[movie_df["title"]==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(enumerate(distance), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(movie_df.iloc[i[0]].title)
       
    
```

```python
recommendation_system("Tom Jones")
```
