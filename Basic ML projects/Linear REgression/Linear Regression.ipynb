```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv("NLP PROJECTS/headbrain.csv")
```

```python
data1.head(5)
```

```python
# let x be headsize
# let y be brain_weight
X=data1["Head Size(cm^3)"].values
```

```python

Y=data1["Brain Weight(grams)"].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```python
print(mean_y)
```

```python
#finding values of "m"(slope) and "c"(intercept)
num=0
den=0
l=len(X)
for i in range(l):
    num+=(X[i]-mean_x)*(Y[i]-mean_y)
    den+=(X[i]-mean_x)**2
# print(num,den)
m_slope=num/den
c_intercept=mean_y-(m_slope*mean_x)
print(m_slope,c_intercept)
```

```python
minx=np.min(X)-100
maxx=np.max(X)+100
x=np.linspace(minx,maxx,500)
y=(m_slope*x)+c_intercept
plt.figure(figsize=(12,6))
plt.plot(x,y, color="red", label='regression line')
plt.scatter(X,Y, color="blue", label="scatter points")
plt.title("headbrain-Head Size(cm^3) Vs Brain Weight(grams)")
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.legend()

```

```python
#using R squared method to check how well the line fits the observations
f_sum_error=0
sec_sum_error=0
for i in range(l):
    y_pred=c_intercept+(m_slope*X[i])
    f_sum_error+=Y[i]-y_pred
    sec_sum_error+=Y[i]-mean_y
R_squared=1-(f_sum_error/sec_sum_error)
print(R_squared)
```
