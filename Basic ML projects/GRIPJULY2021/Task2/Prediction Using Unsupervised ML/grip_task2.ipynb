# 5. Unsupervised Clustering 

# i. Kmeans Clustering

#### Divides data points into k distinct groups.

```python
from sklearn.cluster import KMeans
```

```python
new_iris=iris.iloc[:,[1,3,4]].values
```

```python
# WCSS is nothing but within cluster sum of square 
#kmeans.inertia_ returns us with the wcss value.
wcss_value={}
for i in range(1,10):
    kmeans=KMeans(n_clusters=i, init='k-means++', max_iter=300,random_state=0)
    kmeans.fit(new_iris)
    wcss_value[i]=kmeans.inertia_

print(wcss_value)
```

```python
#the elbow method 
plt.figure(figsize=(8,6))
plt.plot(range(1,10), wcss_value.values(), color="red", marker="o", linewidth=2.5)
plt.title("The Elbow Method\n")
plt.xlabel("No. of Clusters")
plt.ylabel("WSCC value")
plt.show()
```

#### we choose our cluster value as the value where the slope is less steep. Here in this case, i'll be choosing value=3
#### After 3, the drop is also minimal.

```python
kmeans=KMeans(n_clusters=3,
              init='k-means++',
              max_iter=300,
              random_state=0 )
species_predict=kmeans.fit_predict(new_iris)
```

```python
species_predict
```

```python
#visualizing the clusters once between sepallength and petal length
plt.figure(figsize=(12,6))
plt.scatter(new_iris[species_predict==0,0],new_iris[species_predict==0,1], color="cyan", s=30, label="Iris-setosa", marker="h")
plt.scatter(new_iris[species_predict==1,0],new_iris[species_predict==1,1], color="yellow", s=30, label="Iris-versicolour",marker="D" )
plt.scatter(new_iris[species_predict==2,0],new_iris[species_predict==2,1], color="green", s=30, label="Iris-virginica" , marker="p")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="red",s=50, label="centroids")
plt.legend()
plt.show()
```


# ii. Hierarchical Clustering Analysis


#### Subdivides each group into smaller groups

```python
#### Hierarchical Clustering
import scipy
from scipy.cluster.hierarchy import dendrogram,fcluster, linkage
```

```python
iris_HR=iris.iloc[:,[1,3,4]]
iris_HR.index=iris.iloc[:,[5]]
```

```python
iris_HR
```

```python
linkage_matrix=linkage(iris_HR, "ward")
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(100,50))
dendrogram(linkage_matrix, labels=iris_HR.index, leaf_rotation=90., leaf_font_size=20., show_contracted=True)
plt.axhline(y=4)
plt.axhline(y=15)
plt.title("\n\nCluster in the Data\n\n")
plt.xlabel("Species")
plt.ylabel("Distance or Height")
plt.show()
```
